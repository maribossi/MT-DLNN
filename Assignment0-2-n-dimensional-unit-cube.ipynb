```python
import numpy as np
import math
import matplotlib.pyplot as plt

def corners(n):
    return 2**n

#based on sqrt{n*a^2}
def diag(n, r=2):
    return math.sqrt(n*(r**2))

#https://en.wikipedia.org/wiki/Volume_of_an_n-ball
def ball_vol(n, r=1):
    return math.pi**(n/2)/ math.gamma((n/2)+1)
    
def cube_vol(n, r=2):
    return r**n

#number of vertices for dimensions 1-100
dimensions = list(range(1, 100))
vertices = []
for i in dimensions:
    vertices.append(corners(i))

plt.title('n-cube vertices, dimensions 1 to 100')    
plt.xlabel('Dimension')
plt.ylabel('Vertices')  
plt.plot(dimensions, vertices, 'g-')
```

```python
# ball(r=1) volumes for dimensions 1-100
ball_volumes=[]
for i in dimensions:
    ball_volumes.append(ball_vol(i))

plt.xlabel('Dimension')
plt.ylabel('Ball V')  
plt.plot(dimensions, ball_volumes, 'g-')
```

```python
# cube(r=2) volumes for dimensions 1-100   
cube_volumes=[]
for i in dimensions:
    cube_volumes.append(cube_vol(i))

plt.xlabel('Dimension')
plt.ylabel('Cube V')  
plt.plot(dimensions, cube_volumes, 'g-')
```
